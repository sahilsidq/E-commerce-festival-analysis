## ðŸ§¹ Missing Value Handling

```python
print('Missing values count in each column-\n',festf.isnull().sum())
```

```python
unique_event=festf['Event'].value_counts()
print('->No. of unique values of Event:',unique_event)

unique_storeID=festf['StoreID'].value_counts()
print('->No. of unique values of Storeid:',unique_storeID)

unique_shippingMethod=festf['ShippingMethod'].value_counts()
print('->No. of unique values of shipping method:',unique_shippingMethod)

unique_deliveryTime=festf['DeliveryTime'].value_counts()
print('->No. of unique values of shipping method:',unique_deliveryTime)
```

```python

festf['Event']=festf['Event'].fillna(festf['Event'].mode()[0])
festf['ShippingMethod']=festf['ShippingMethod'].fillna(festf['ShippingMethod'].mode()[0])
festf.drop(columns=['StoreID'],inplace=True)
print("Deleted StoreId Column Successfully")
festf.drop(columns=['DeliveryTime'],inplace=True)
print("Deleted DeliveryTime Column Successfully")
```

<li style=color:blue><i><b>Deleting columns StoreId and DeliveryTime cause not  necessary for the analysis</b></i></li> 
<li style=color:blue><i><b>Replacing the missing values with most frequent one in columns Event and Shippingmethod cause is necessary for analysis</i></b></li>

# ðŸ“Š Visualizations

```python
num_col=festf.select_dtypes(include=['int64','float64'])
cor_mat=num_col.corr()
#visualization of correlation matrix using HeatMap
sns.heatmap(cor_mat,annot=True,cmap='coolwarm',fmt='.2f',linewidths=0.5)
print(cor_mat)
plt.title('Correlation Heatmap')
```

<p style=color:blue><i><b>This correlation matrix visualizes the relationships between different features in the Festive sales dataset. The values range from -1 to 1, where:</b></i></p>
<li style=color:blue><i> 1 indicates a perfect positive correlation.</i></li>
<li style=color:blue><i> -1 indicates a perfect negative correlation.</i></li>
<li style=color:blue><i> 0 indicates no correlation.</i></li>

**************************************************************************************************************************************

```python
plt.figure(figsize=(10,6))
sns.histplot(festf['TotalPrice'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Total Sales')
plt.xlabel('Total Price')
plt.ylabel('Frequency')
plt.show()
```

<h3>This Plot Shows :</h3>
<li style=color:blue><i><b>Most transactions have moderate total sales, but a few are significantly high.</b></i></li>

***************************************************************

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Gender', y='TotalPrice', data=festf, estimator=sum,palette="Paired",hue='Gender',legend=False)
plt.title('Total Sales by Gender')
plt.ylabel('Total Sales')
plt.show()
```

<h3>This Plot Shows :</h3>
<li style=color:blue><i><b>One gender segment consistently contributes more to overall festive sales.</b></i></li>

**********************************************************************************************************

```python

sales_by_festival = festf.groupby('Category')['TotalPrice'].sum()

plt.figure(figsize=(8,8))
plt.pie(sales_by_festival, labels=sales_by_festival.index, autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Festival')
plt.show()
```

<h3>This Plot Shows :</h3>
<li style=color:blue><i><b>Some product categories yield higher average sales per transaction as you can see in the chart most sales are from "Toys" and "Food" categories. </b></i></li>

*************************************************************************************************

```python
plt.figure(figsize=(8,5))
sns.countplot(x='CustomerSatisfaction', data=festf, palette='coolwarm',hue='CustomerSatisfaction',legend=False)
plt.title('Customer Satisfaction Ratings')
plt.show()
```

<h3>This Plot Shows :</h3>
<li style=color:blue><i><b>Most users gives neutral satisfaction, showing general approval of service/product.</b></i></li>

****************************************

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='DiscountAmount', y='TotalPrice', data=festf, alpha=0.6)
plt.title('Sales vs Discount')
plt.show()
```

<h3>This Plot Shows :</h3>
<li style=color:blue><i><b>Discounts can help increase sales.</b></i></li>

************************************************

```python
return_rate = festf.groupby('Category')['ReturnFlag'].apply(lambda x: (x=='True').mean()).sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.boxplot(x=return_rate.index, y=return_rate.values, )
plt.title('Return Rate by Category')
plt.ylabel('Return Rate')
plt.xticks(rotation=45)
plt.show()
```

<h3>This Plot Shows :</h3>
<li style=color:blue><b><i>Return Rate by different categories and rate of return are <b>same</b> in all the category.</i></b></li>

*************************************************************************************************

```python
top_cities = festf['Location'].value_counts().nlargest(10).index
plt.figure(figsize=(10,6))
sns.barplot(x='Location', y='TotalPrice', data=festf[festf['Location'].isin(top_cities)], estimator=sum,palette='rocket',legend=False,hue='Location')
plt.title('Total Sales in Top 10 Locations')
plt.xticks(rotation=45)
plt.show()
```

<h3>This Plot Shows :</h3>
<li style=color:blue><b><i>Most orders are from which city like "city 13" has most orders.</i></b></li>
